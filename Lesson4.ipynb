```python
# 2. Посчитайте, сколько целых месяцев длилась добыча на каждой скважине в файле wells_info.csv.   
import numpy as np
import pandas as pd
import numpy as np

wells_info_df = pd.read_csv('data/wells_info.csv')
    
wells_info_df['CompletionDate'] = pd.to_datetime(wells_info_df['CompletionDate'])
wells_info_df['PermitDate'] = pd.to_datetime(wells_info_df['PermitDate'])

init_year = wells_info_df['PermitDate'].dt.year
final_year = wells_info_df['CompletionDate'].dt.year

init_month = wells_info_df['PermitDate'].dt.month
final_month = wells_info_df['CompletionDate'].dt.month

init_day = wells_info_df['PermitDate'].dt.day
final_day = wells_info_df['CompletionDate'].dt.day

days = init_day - final_day
    
dif = (final_year - init_year) * 12 + (final_month - init_month) - np.heaviside(days, 0)

number_of_months = pd.DataFrame( dif , columns = ['Number of months'])
df = pd.concat([wells_info_df['PermitDate'], wells_info_df['CompletionDate'], number_of_months], axis=1)
display(df)
```

```python
# 4. Используя файл production.csv добавьте к каждой скважине в файле wells_info.csv  
# колонку с информацией о суммарной добыче за всё время и за первые 12 месяцев.
import numpy as np
import pandas as pd

production_df = pd.read_csv('data/production.csv')
wells_info_df = pd.read_csv('data/wells_info.csv')

# За все время
df_all_liquid = production_df.groupby('API').sum()[['Liquid']].sum(axis = 1)
df_all_gas = production_df.groupby('API').sum()[['Gas']].sum(axis = 1)
df_all_water = production_df.groupby('API').sum()[['Water']].sum(axis = 1)

df_sum_liquid = pd.DataFrame(df_all_liquid, columns = ['Liquid production'])
df_sum_gas = pd.DataFrame(df_all_gas, columns = ['Gas production'])
df_sum_water = pd.DataFrame(df_all_water, columns = ['Water production'])
df_all = pd.merge(wells_info_df, df_sum_liquid, on='API')
df_all = pd.merge(df_all, df_sum_gas, on='API')
df_all = pd.merge(df_all, df_sum_water, on='API')

# За 12 месяцев
groups = pd.DataFrame(production_df.groupby('API').head(12))

df_12m_liquid = groups.groupby('API').sum()[['Liquid']].sum(axis = 1)
df_12m_liquid = pd.DataFrame(df_12m_liquid,columns = ['Liquid production(12 months)'])

df_12m_gas = groups.groupby('API').sum()[['Gas']].sum(axis = 1)
df_12m_gas = pd.DataFrame(df_12m_gas,columns = ['Gas production(12 months)'])

df_12m_water = groups.groupby('API').sum()[['Water']].sum(axis = 1)
df_12m_water = pd.DataFrame(df_12m_water,columns = ['Water production(12 months)'])

df = pd.merge(df_all, df_12m_liquid, on='API')
df = pd.merge(df, df_12m_gas, on='API')
df = pd.merge(df, df_12m_water, on='API')

# Вывод
compression_opts = dict(method='zip', archive_name='out.csv') 
df.to_csv('out.zip', index=False, compression=compression_opts)
```
